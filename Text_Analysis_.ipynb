## Data Cleaning

```python
#lets define a function to clean our text.
def clean(text):
    
    #removing editor's information
    x = re.search("Blackcoffer Insights", text)
    if x != None:
        a=x.span()
        text = text[:a[0]-1]
        
    #removing Share: from end
    text = re.sub("Share:",'',text)
    
    #removing urls
    text = re.sub("^http\://[a-zA-Z0-9-\.]+\.[a-zA-Z]{2,3}(/\S*)?$", '',text)
    #remove white space
    text = text.strip()
    
    return(text)
```

```python
#lets clean the text
df['Article'] = df['Article'].apply(lambda x : clean(x))
```

```python
#lets check our first article
df.Article[0]
```

### Before removing stopwords from the text we will find the count for Personal Pronouns in the text and create a column for it.

```python
def pronoun_counter(text):
    words=word_tokenize(text)
    count=0
    for i in words:
        if i in ["I","we","my","ours","us"]:
            count+=1
    return(count)
```

```python
#lets clean the text
df['PRONOUN COUNT'] = df['Article'].apply(lambda x : pronoun_counter(x))
```

## Removing stop_words from the text

### As assisted in problem statement, I will load file from given link contaning list for stopwords

```python
#loading the file
file = open("StopWords_GenericLong.txt",'r')
stop_words = file.read().split('\n')

#removing stopwords from text
df['Article'] = df['Article'].apply(lambda x : ' '.join([word for word in x.lower().split() if word not in (stop_words)]))
```

We will go for calculation of positive and negative score. For that we will create lists of positive and negative words, for this task I have used word dictionary that is mentioned in our problem statement.

```python
#getting Master Dictionary
Word_dict = pd.read_csv(r"Loughran-McDonald_MasterDictionary_1993-2021.csv")
Word_dict.head(5)
```

```python
Word_dict.shape
```

```python
#Lets create lists for positive and negative words
positive=[]
negative = []
for idx in range(Word_dict.shape[0]):
    if Word_dict.Positive[idx]!=0:
        positive.append(Word_dict.Word[idx].lower())
    if Word_dict.Negative[idx]!=0:
        negative.append(Word_dict.Word[idx].lower())
```

```python
#lets check the content of our lists
print("Number of words in positive list: ", len(positive))
print("\nNumber of words in negative list: ", len(negative))
print("\nFirst 10 positive words in the positive list: ", positive[0:10])
print("\nFirst 10 negative words in the negative list: ", negative[0:10])
```

Great! now we are having two lists for postive and negative words.

```python
#lets create a function to calculate positive score for text

#first we will lower all text
df["Article"]=df["Article"].apply(lambda x: x.lower())

#defining a function to calculate positive score
def pos_score(text):
    count=0
    word_token = word_tokenize(text)
    for i in word_token:
        if i in positive:
            count+=1
    return count

#defining a function to calculate negative score
def neg_score(text):
    count=0
    word_token = word_tokenize(text)
    for i in word_token:
        if i in negative:
            count-=1
    return count*(-1)
```

```python
df["POSITIVE SCORE"]=df["Article"].apply(lambda x: pos_score(x))
df["NEGATIVE SCORE"]=df["Article"].apply(lambda x: neg_score(x))
```

## Analysis of Readability

To analyse readability we will find Average Sentence Length, Percentage of Complex words and fog index

```python
#Average sentence length
def text_cleaner(text):
    #removeing punctuations from text
    txt = re.sub(r"[^\w\s]", "", text)
    #remove html markup
    txt = re.sub("(<.*?>)", "", txt)
    #remove non-ascii and digits
    txt = re.sub("(\\W)", " ", txt)
    txt = re.sub("(\\d)", "", txt)
    return txt

def avg_sent_len(text):
    sentence = sent_tokenize(text)
    txt = text_cleaner(text)
    words = word_tokenize(txt)
    
    return round((len(words)/len(sentence)),3)

def Word_count(text):
    #lets clean the text first
    txt = text_cleaner(text)
    words = word_tokenize(txt)
    return len(words)

#funftion to calculate syllable for a word
def syllable_count(word):
    count = 0
    vowels = ['a','e','i','o','u']
    if word[0] in vowels:
        count += 1
    for index in range(1, len(word)):
        if word[index] in vowels and word[index - 1] not in vowels:
            count += 1
            if word.endswith("e") or word.endswith('ed') or word.endswith('es'):
                count -= 1
    if count == 0:
        count += 1
    return count

#function to calculate number of complex words per text
def Count_complex_words(text):
    txt = text_cleaner(text) #text cleaning
    words = word_tokenize(txt)
    count = 0
    for i in words:
        if syllable_count(i)>2:
            count+=1
    return count

#function to calculate syllable per word
def Sl_per_word(text):
    txt = text_cleaner(text) #text cleaning
    words = word_tokenize(txt)
    sl_count=0
    #loop to count all syllables from text
    for i in words:
        syl =syllable_count(i)
        sl_count+=syl
    return round((sl_count/len(words)),3)

#function to calculate average word length
def Avg_word_len(text):
    txt = text_cleaner(text) #text cleaning
    words = word_tokenize(txt)
    word_len = 0
    for wrd in words:
        word_len+=len(wrd)
    return round((word_len/len(words)),3)


#function to calculate Average Number of Words Per Sentence
def words_per_sentence(text):
    sentence = sent_tokenize(text)
    txt = text_cleaner(text)
    words = word_tokenize(txt)
    return round((len(words)/len(sentence)),3)
```

```python
#Average sentence length
df["AVERAGE SENTENCE LENGTH"]=df['Article'].apply(lambda x: avg_sent_len(x))

#calculating number complex words
df["COMPLEX WORD COUNT"]=df['Article'].apply(lambda x: Count_complex_words(x))

#word count
df["WORD COUNT"]=df['Article'].apply(lambda x: Word_count(x))

#Count of sellables per word
df["SYLLABLE PER WORD"] = df['Article'].apply(lambda x: Sl_per_word(x))

#Percentage of complex words
df["PERCENTAGE OF COMPLEX WORDs"] = round(df["COMPLEX WORD COUNT"]/df["WORD COUNT"],3)

#fog index
df["FOG INDEX"] = round(0.4*(df['AVERAGE SENTENCE LENGTH']+df["PERCENTAGE OF COMPLEX WORDs"]),3)

#Average number of words per sentence
df["AVERAGE NUMBER OF WORDS PER SENTENCE"] = df['Article'].apply(lambda x: words_per_sentence(x))

#Average Word Length
df["AVERAGE WORD LENGTH"] = df['Article'].apply(lambda x: Avg_word_len(x))
```

## Calculate Polarity and Subjectivity scores

As the formulae for calculation polarity and subjectivity of the text are given, we will use those and calculate the same.

```python
#Polarity calculation
df["POLARITY SCORE"] = (df["POSITIVE SCORE"]-df["NEGATIVE SCORE"])/((df["POSITIVE SCORE"]+df["NEGATIVE SCORE"])+0.000001)

#Subjectivity calculation
df["SUBJECTIVITY SCORE"] = (df["POSITIVE SCORE"]+df["NEGATIVE SCORE"])/(df['WORD COUNT']+0.000001)
```
